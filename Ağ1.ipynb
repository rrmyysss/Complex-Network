```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Boş bir yönsüz ağ oluştur
G=nx.Graph()

# Düğümleri ekle
G.add_nodes_from(["A","B","C","D"])

# Kenarları ekle, yani düğümleri birbirine bağladık
G.add_edges_from([("A","B"),("B","C"),("C","D"),("D","A")])
```

```python
# Basit bir yönsüz ağ
nx.draw(G,with_labels=True,node_color='skyblue',node_size=1500,font_size=16)
plt.title("Basit Yönsüz Ağ")
plt.show()
```

```python
# Basit bir yönlü ağ
DG=nx.DiGraph()
DG.add_edges_from([("A","B"),("B","C"),("C","A"),("C","D")]) # ilk hangisi yazıldıysa ondan diğerine ok yönü belirlenir

plt.figure(figsize=(8,6))
nx.draw(DG,with_labels=True,node_color='lightgreen',node_size=1200,arrows=True)
plt.title("Basit Yönlü Ağ")
plt.show()
```

```python
# Basit bir ağırlıklı graf
WG=nx.Graph()

# (düğüm1, düğüm2, ağırlık)
WG.add_weighted_edges_from([("A","B",4.5),("B","C",2.0),("C","D",1.2),("D","A",3.8)])

pos = nx.spring_layout(WG) # Konum hesaplama
weights = nx.get_edge_attributes(WG,'weight') # Ağırlıkları al

plt.figure(figsize=(5,5))
nx.draw(WG,pos,with_labels=True,node_color='lightcoral',node_size=1200)
nx.draw_networkx_edge_labels(WG,pos,edge_labels=weights)
plt.title("Ağırlıklı Ağ")
plt.show()
```

## Topoloji
Ağ topolojisi, bir bilgisayar ağındaki cihazların (düğümlerin) ve bağlantıların (bağlantıların) düzenlenmesidir. Bilgisayarların, sunucuların ve diğer cihazların nasıl bağlandığını ve aralarında nasıl veri akışı olduğunu gösterir. İki ana topoloji türü vardır:
*   Fiziksel Topoloji: Kabloların ve cihazların gerçek fiziksel düzeni.
*   Mantıksal Topoloji: Verilerin fiziksel düzenden bağımsız olarak ağ üzerinde nasıl hareket ettiği.



## Topoloji Çeşitleri

1.   Noktadan Noktaya Topoloji: Birinin gönderici birinin alıcı olduğu iki düğüm arasındaki en basit iletişimdir.
2.   Ağ Topolojisi : Her cihaz özel kanallar aracılığıyla başka bir cihaza bağlanır.
3. Yıldız Topolojisi : Tüm cihazlar bir kablo aracılığıyla tek bir merkezi düğüme bağlanır.
4. Otobüs Topolojisi : Tüm ağ cihazları tek bir kabloya bağlıdır. Çift yönlüdür.
5. Halka Topolojisi : Her cihaz için iki komşu cihazla bağlantı kurularak halka oluşturulur.
6. Ağaç Topolojisi : Ağaç yapısında hiyerarşik veri akışı vardır.
7. Hibrit Topoloji : Tüm topolojilerin birleşimidir.




```python
# Topolojileri çizmek için fonksiyon
def draw_graph(G, title):
    pos = nx.spring_layout(G, seed=42)
    plt.figure(figsize=(6,4))
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, edge_color='gray', font_size=14, font_weight='bold')
    plt.title(title, size=16)
    plt.show()
```

```python
# 1. Noktadan Noktaya Topoloji
P2P = nx.Graph()
P2P.add_nodes_from(["Gönderici", "Alıcı"]) # Düğümler
P2P.add_edges_from([("Gönderici", "Alıcı")]) # Bağlantı

draw_graph(P2P, "Noktadan Noktaya Topoloji")
```

```python
# 2. Ağ (Mesh) Topolojisi
M = nx.complete_graph(5)

# Düğümleri daha anlamlı isimlendirelim (isteğe bağlı)
mapping = {0: "1", 1: "2", 2: "3", 3: "4", 4: "Sunucu"}
MESH = nx.relabel_nodes(M, mapping)

draw_graph(MESH, "Ağ (Mesh) Topolojisi")
```

```python
# 3. Yıldız Topolojisi
STAR = nx.Graph()

# Merkezi düğüm
merkezi_dugum = "Switch"
STAR.add_node(merkezi_dugum)

# Çevresel cihazları ekleyip merkeze bağlayalım.
cihazlar = ["PC-A", "PC-B", "Yazıcı", "Sunucu"]
for cihaz in cihazlar:
    STAR.add_node(cihaz)
    STAR.add_edge(merkezi_dugum, cihaz)

pos = nx.spring_layout(STAR, seed=42)
# Merkezi düğümü ortaya alarak daha iyi bir görsel elde edelim
pos[merkezi_dugum] = [0, 0]
plt.figure(figsize=(6,4))
nx.draw(STAR, pos, with_labels=True, node_color='skyblue', node_size=1500, edge_color='gray', font_size=12, font_weight='bold')
plt.title("Yıldız Topolojisi", size=15)
plt.show()
```

```python
# 4. Otobüs Topolojisi
BUS = nx.path_graph(7) # 5 cihaz + 2 sonlandırıcı

# Düğümler
mapping = {
    0: "Sonlandırıcı-1",
    1: "PC-1",
    2: "PC-2",
    3: "PC-3",
    4: "Yazıcı",
    5: "Sunucu",
    6: "Sonlandırıcı-2"
}
BUS = nx.relabel_nodes(BUS, mapping)

# Grafiği çizdiriyoruz.
pos = nx.spring_layout(BUS, seed=42)
plt.figure(figsize=(10, 4))
nx.draw(BUS, pos, with_labels=True, node_color=['red'] + ['skyblue']*5 + ['red'], node_size=1500, edge_color='black', width=3.0, font_size=10)
plt.title("Otobüs Topolojisi", size=15)
plt.show()
```

```python
# 5. Halka Topolojisi
RING = nx.cycle_graph(6)

# Düğümleri isimlendir
mapping = {0: "İstasyon A", 1: "İstasyon B", 2: "İstasyon C", 3: "İstasyon D", 4: "İstasyon E", 5: "İstasyon F"}
RING = nx.relabel_nodes(RING, mapping)

# Grafiği çizdiriyoruz.
# Halka yapısını en iyi gösteren çizim düzeni "circular_layout"tur.
pos = nx.circular_layout(RING)
plt.figure(figsize=(6,4))
nx.draw(RING, pos, with_labels=True, node_color='skyblue', node_size=1800, edge_color='gray', font_size=12, font_weight='bold')
plt.title("Halka Topolojisi", size=15)
plt.show()
```

```python
# 6. Ağaç Topolojisi
# r: her düğümün dal sayısı (branching factor)
# h: ağacın yüksekliği (height)
TREE = nx.balanced_tree(r=2, h=3) # 2 dal, 3 yükseklik

# Hiyerarşik yapıyı göstermek için özel bir pozisyonlama gerekebilir.
# "graphviz" ile daha iyi sonuç alınabilir ama basit bir gösterim için spring_layout yeterli.
pos = nx.nx_pydot.graphviz_layout(TREE, prog='dot') # graphviz düzeni hiyerarşiyi daha iyi gösterir.

plt.figure(figsize=(10, 4))
nx.draw(TREE, pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=10)
plt.title("Ağaç Topolojisi", size=15)
plt.show()
```

```python
# 7. Hibrit Topoloji
HYBRID = nx.Graph()

# Birinci Yıldız Topolojisi
merkez1 = "Switch-1"
HYBRID.add_node(merkez1)
for i in range(1, 4):
    cihaz = f"PC-A{i}"
    HYBRID.add_node(cihaz)
    HYBRID.add_edge(merkez1, cihaz)

# İkinci Yıldız Topolojisi
merkez2 = "Switch-2"
HYBRID.add_node(merkez2)
for i in range(1, 4):
    cihaz = f"PC-B{i}"
    HYBRID.add_node(cihaz)
    HYBRID.add_edge(merkez2, cihaz)

# İki merkezi düğümü (switch'leri) birbirine bağlayarak hibrit yapıyı oluşturalım.
HYBRID.add_edge(merkez1, merkez2)


# Grafiği çizdiriyoruz.
draw_graph(HYBRID, "Hibrit Topoloji (Yıldız-Yıldız)")
```

# Hazır fonksiyon kullanmadan Topolojiler

```python
class BasitGraf:
    #Veriyi bir komşuluk listesi (sözlük) içinde tutar. Örnek: {'A': ['B', 'C'], 'B': ['A'], 'C': ['A']}

    def __init__(self):
        self.komsu_listesi = {}

    def dugum_ekle(self, dugum):
        #Grafiğe bir düğüm ekler. Düğüm zaten varsa bir şey yapmaz. Düğümün komşu listesini boş bir liste olarak başlatır.
        if dugum not in self.komsu_listesi:
            self.komsu_listesi[dugum] = []

    def kenar_ekle(self, dugum1, dugum2):
        #İki düğüm arasına yönsüz bir kenar ekler. Önce düğümlerin var olduğundan emin olur, yoksa ekler.

        # 1. Düğümlerin var olduğundan emin ol
        self.dugum_ekle(dugum1)
        self.dugum_ekle(dugum2)

        # 2. Yönsüz olduğu için bağlantıyı iki taraflı ekle. (Eğer bağlantı zaten varsa tekrar eklemez)
        if dugum2 not in self.komsu_listesi[dugum1]:
            self.komsu_listesi[dugum1].append(dugum2)

        if dugum1 not in self.komsu_listesi[dugum2]:
            self.komsu_listesi[dugum2].append(dugum1)

    def goster(self, title):
        """
        Grafiğin mevcut yapısını (komşuluk listesini) başlıkla birlikte yazdırır.
        """
        print(f"\n--- {title} ---")
        print("Graf Yapısı (Komşuluk Listesi):")

        if not self.komsu_listesi:
            print("  (Graf boş)")
            print("------------------------" + "-" * len(title))
            return

        # Daha düzenli bir çıktı için
        max_len = max(len(d) for d in self.komsu_listesi.keys())

        for dugum, komsular in self.komsu_listesi.items():
            komsular_str = ", ".join(komsular) if komsular else "Yok"
            # Düğüm adını hizalayarak yazdır
            print(f"  Düğüm '{dugum}':".ljust(max_len + 10) + f" Komşuları -> [{komsular_str}]")
        print("------------------------" + "-" * len(title))
```

```python
# --- 1. Noktadan Noktaya Topoloji ---
P2P = BasitGraf()
P2P.kenar_ekle("Gönderici", "Alıcı")
P2P.goster("Noktadan Noktaya Topoloji")


# --- 2. Ağ (Mesh) Topolojisi ---
Mesh = BasitGraf()
cihazlar = ["Cihaz 1", "Cihaz 2", "Cihaz 3", "Cihaz 4"]

# Her cihazı (i) kendisinden sonraki (j) her cihaza bağlayan döngü
for i in range(len(cihazlar)):
    for j in range(i + 1, len(cihazlar)):
        Mesh.kenar_ekle(cihazlar[i], cihazlar[j])

Mesh.goster("Ağ (Mesh) Topolojisi")


# --- 3. Yıldız Topolojisi ---
Star = BasitGraf()
merkezi_hub = "Merkezi Hub"
uclar = ["PC 1", "PC 2", "PC 3", "PC 4", "Yazıcı"]

# Tüm uç cihazları tek tek merkeze bağla
for uc in uclar:
    Star.kenar_ekle(merkezi_hub, uc)

Star.goster("Yıldız Topolojisi")


# --- 4. Otobüs Topolojisi ---
Bus = BasitGraf()
bus_sirasi = ["Cihaz A", "Cihaz B", "Cihaz C", "Cihaz D"]

# Cihazları sırayla birbirine bağla (A->B, B->C, C->D)
for i in range(len(bus_sirasi) - 1):
    Bus.kenar_ekle(bus_sirasi[i], bus_sirasi[i+1])

Bus.goster("Otobüs Topolojisi")


# --- 5. Halka Topolojisi ---
Ring = BasitGraf()
halka_cihazlari = ["Cihaz 1", "Cihaz 2", "Cihaz 3", "Cihaz 4"]
n = len(halka_cihazlari)

for i in range(n):
    # 'i' indeksli cihazı 'i+1' indeksli cihaza bağla
    # Modulo (%) operatörü, son cihazın (n-1) ilk cihaza (0) bağlanmasını sağlar.
    dugum1 = halka_cihazlari[i]
    dugum2 = halka_cihazlari[(i + 1) % n]
    Ring.kenar_ekle(dugum1, dugum2)

Ring.goster("Halka Topolojisi")


# --- 6. Ağaç Topolojisi ---
Tree = BasitGraf()
Tree.kenar_ekle("Kök Sunucu", "Ana Switch")
Tree.kenar_ekle("Ana Switch", "Switch A")
Tree.kenar_ekle("Ana Switch", "Switch B")
Tree.kenar_ekle("Switch A", "PC 1")
Tree.kenar_ekle("Switch A", "PC 2")
Tree.kenar_ekle("Switch B", "PC 3")
Tree.kenar_ekle("Switch B", "Yazıcı")

Tree.goster("Ağaç Topolojisi")


# --- 7. Hibrit Topoloji ---
Hybrid = BasitGraf()

# 1. Yıldız (Hub A)
Hybrid.kenar_ekle("Hub A", "PC 1")
Hybrid.kenar_ekle("Hub A", "PC 2")

# 2. Yıldız (Hub B)
Hybrid.kenar_ekle("Hub B", "PC 3")
Hybrid.kenar_ekle("Hub B", "PC 4")

# 3. İki Hub'ı bir otobüs hattı (bir kenar) ile birleştir
Hybrid.kenar_ekle("Hub A", "Hub B")

Hybrid.goster("Hibrit Topoloji (Star-Bus-Star)")
```
