**Hazır fonksiyon kullanmadan**

1) Degree Centrality

Bir düğümün derece merkeziliği = (Düğümün derecesi) / (Grafikteki toplam düğüm sayısı - 1)


```python
# 1. Toplam düğüm sayısını al
num_nodes = G.number_of_nodes()

# 2. Merkezilikleri saklamak için bir sözlük (dictionary) oluştur
manual_degree_centrality = {}

# 3. Normalizasyon için kullanılacak böleni hesapla (N-1)
# Eğer graf 1 veya 0 düğümden oluşuyorsa, ZeroDivisionError almamak için böleni 1 yapıyoruz. (Merkezilik zaten 0 olacaktır)
if num_nodes <= 1:
    normalization_factor = 1.0
else:
    normalization_factor = float(num_nodes - 1)

# 4. Her düğüm için merkeziliği manuel olarak hesapla
for node in G.nodes():
    degree = G.degree(node)
    centrality = degree / normalization_factor
    # Sonucu sözlüğe ekle
    manual_degree_centrality[node] = centrality
```

2) Closeness Centrality

Bir $u$ düğümünün yakınlık merkeziliği = ( $u$ 'dan ulaşılabilen düğüm sayısı - 1) / ( $u$ 'dan diğer tüm ulaşılabilen düğümlere olan en kısa yol uzunluklarının toplamı)

```python
# 1. Merkezilikleri saklamak için bir sözlük (dictionary) oluştur
manual_closeness_centrality = {}

# 2. Her düğüm için yakınlık merkeziliğini manuel olarak hesapla
for node in G.nodes():

    # nx.shortest_path_length, bir düğümden diğer tüm *ulaşılabilir* düğümlere olan en kısa yol uzunluklarını bir sözlük olarak verir.
    # {hedef_düğüm: mesafe, ...} Örn: node=2 için -> {2: 0, 1: 1, 3: 1, 4: 1, 5: 2}
    shortest_paths = nx.shortest_path_length(G, source=node)

    # Mesafelerin toplamını al (payda) .values() mesafeleri (0, 1, 1, 1, 2) verir
    total_distance = sum(shortest_paths.values())

    # Ulaşılabilen düğüm sayısını al (pay)
    # shortest_paths sözlüğü kaynak düğümü de içerir (mesafe 0 olarak). Formül "diğer" düğümleri istediği için toplam sayıdan 1 çıkarırız.
    num_reachable_nodes = len(shortest_paths) - 1

    # 4. Merkeziliği hesapla ve kenar durumları kontrol et
    if total_distance == 0:
        # Eğer toplam mesafe 0 ise (düğüm izoledir, kimseye ulaşamaz), merkezilik 0 olarak tanımlanır.
        manual_closeness_centrality[node] = 0.0
    else:
        # Formülü uygula
        centrality = num_reachable_nodes / total_distance
        manual_closeness_centrality[node] = centrality
```

3. Betweennes Centrality


Bir düğümün aradalık merkeziliği, o düğümün, grafikteki diğer düğüm çiftleri arasındaki en kısa yollar üzerinde ne sıklıkla bir "köprü" veya "aracı" görevi gördüğünü ölçer.

```python
nodes = G.nodes()
manual_betweenness_centrality = {node: 0.0 for node in nodes}

# 1. Tüm benzersiz düğüm çiftlerini (s, t) al
# itertools.combinations(nodes, 2) -> (1,2), (1,3), (1,4), (1,5), (2,3), ...
all_pairs = list(itertools.combinations(nodes, 2))

# 2. Her bir düğüm çifti (s, t) için hesaplama yap
for s, t in all_pairs:

    # s ve t arasındaki TÜM en kısa yolları bul
    try:
        all_shortest_paths = list(nx.all_shortest_paths(G, source=s, target=t))
    except nx.NetworkXNoPath:
        # Aralarında yol yoksa (graf bağlı değilse) bu çifti atla
        continue

    # Toplam en kısa yol sayısı (sigma_st)
    num_shortest_paths = len(all_shortest_paths)

    if num_shortest_paths == 0:
        continue

    # 3. Bu yolların hangi ara düğümlerden (v) geçtiğini say(Her çift için geçici bir sayaç)
    node_pass_count = {node: 0 for node in nodes}

    for path in all_shortest_paths:
        # s ve t hariç (yani path[1:-1]), aradaki düğümleri al
        intermediate_nodes = path[1:-1]
        for v in intermediate_nodes:
            node_pass_count[v] += 1

    # 4. Her ara düğümün (v) merkezilik skorunu güncelle
    # Katkı = sigma_st(v) / sigma_st
    for v in nodes:
        if v != s and v != t: # Sadece ara düğümlerin skoru artar
            contribution = node_pass_count[v] / num_shortest_paths
            manual_betweenness_centrality[v] += contribution

# 5. Normalizasyon
# networkx, varsayılan olarak (normalized=True) skorları [0, 1] arasına ölçeklemek için bir normalizasyon faktörüne böler.
num_nodes = G.number_of_nodes()
if num_nodes <= 2:
    # 2 veya daha az düğüm varsa normalizasyon anlamsızdır/imkansızdır.
    pass
else:
    # Yönsüz (undirected) graf için normalizasyon faktörü: (N-1)(N-2)/2.Bu, grafikteki tüm olası çiftlerin sayısıdır.
    normalization_factor = ((num_nodes - 1) * (num_nodes - 2)) / 2.0
    for node in manual_betweenness_centrality:
        manual_betweenness_centrality[node] /= normalization_factor
```

4. EigenVector

Bir düğümün önemi, önemli komşulara sahip olmasına bağlıdır.

```python
# 1. Algoritma parametreleri
max_iter = 100  # Sonsuz döngüyü önlemek için maksimum tekrar
tolerance = 1e-6 # Durma koşulu: Değişim bu değerin altına düşerse
nodes = list(G.nodes())
num_nodes = G.number_of_nodes()

if num_nodes == 0:
    print("Graf boş.")
    # Boş bir sözlükle çıkılabilir
    manual_eigenvector_centrality = {}
else:
    # 2. Başlangıç skoru: Tüm düğümlere 1.0 ata
    x = {node: 1.0 for node in nodes}

    # 3. Güç Yöntemi (Power Iteration) döngüsü
    for _ in range(max_iter):
        x_new = x.copy()
        total_error = 0.0

        # 4. Her düğümün skorunu komşularının toplamıyla güncelle
        # (Bu, x_new = A * x matris çarpımıdır)
        for node in nodes:
            score_sum = 0.0
            for neighbor in G.neighbors(node):
                score_sum += x[neighbor] # Komşunun bir önceki adımdaki skoru
            x_new[node] = score_sum

        # 5. Normalizasyon (L2 Norm)
        # Sayıların çok büyümesini engeller
        try:
            norm_sq = sum(v**2 for v in x_new.values())
            norm = math.sqrt(norm_sq)
            if norm == 0:
                # Kenarı olmayan bir graf gibi, tüm skorlar 0'a çöktü
                break
        except OverflowError:
            # Sayılar çok büyüdü (genelde olmaz ama iyi bir kontroldür)
            print("Hata: Skorlar çok büyüdü.")
            break

        for node in nodes:
            x_new[node] /= norm

        # 6. Durma koşulunu (yakınsama) kontrol et
        # Eski ve yeni vektör arasındaki farka bak
        total_error = sum(abs(x_new[node] - x[node]) for node in nodes)

        # Vektörleri bir sonraki iterasyon için güncelle
        x = x_new

        if total_error < num_nodes * tolerance:
            # Değişim çok azaldı, skorlar "duruldu" (yakınsadı)
            break

    # Döngü bittiğinde, 'x' bizim merkezilik değerlerimizi tutar
    manual_eigenvector_centrality = x
```
