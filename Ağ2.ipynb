```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.karate_club_graph()
nx.draw(G,with_labels=True)
plt.show()
```

```python
print("Nodes: ", G.nodes())
print("Edges: ", G.edges())
```

```python
# degree
# G.degree[n] -> dm n'nin derecesini dner
print("Node 0 degree: ", G.degree[0])
print("Node 32 degree: ", G.degree[32])
print("Node 8 degree: ", G.degree[8])
```

```python
# Tm a iin
for node, deg in G.degree():
  print(f"Node {node}: Degree {deg}")
```

```python
# degree_histogram : Adaki dmlerin derece dalmlarn verir(histogram olarak)
hist = nx.degree_histogram(G)
print(hist)
plt.bar(range(len(hist)), hist)
plt.xlabel("Degree")
plt.ylabel("Number of Nodes")
plt.title("Degree Histogram")
plt.show()
```

```python
# density : An younluunu hesaplar. Yani adaki mevcut kenarlarn, olas maksimum kenarlara oran
density = nx.density(G)
print("Density of the graph",density)
# iletiim genelde 2 dm arasnda olduu iin younluk dk olur
```

```python
# number_of_nodes : Grafikteki toplam dm saysn dner
print("Number of nodes: ", nx.number_of_nodes(G))
```

```python
# number_of_edges
print("Number of edges: ", nx.number_of_edges(G))
```

```python
# density hesaplama(ynsz graf) : 2 * (78/(33*34)) = 0.13
```

```python
# neighbours : Belirtilen dmn komularn dner. Bu bir iterator dner.
neighbors = list(G.neighbors(0))
print("Neighbours of node 0 :",neighbors)
```

```python
all_neighbors = list(G.neighbors(5))
print("Neighbours of node 5 :",all_neighbors)
```

```python
# ki dmn ortak komular
common_neighbors = list(nx.common_neighbors(G, 0, 1))
print("Common neighbors of node 0 and 1: ", common_neighbors)
```

```python
# subgraph
nodes = [0,1,6,10]
SG = G.subgraph(nodes)
nx.draw(SG,with_labels=True)
plt.title("Subgraph of nodes")
plt.show()
```
