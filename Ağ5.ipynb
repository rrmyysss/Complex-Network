```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
edges_csv = pd.read_csv("/content/InputFileEdges.csv")
nodes_csv = pd.read_csv("/content/InputFileNodes.csv")
```

```python
nodes_csv = nodes_csv.rename(columns={"id": "node"})
nodes = nodes_csv["node"]
```

```python
edges = edges_csv[['from', 'to']].apply(tuple, axis=1).unique().tolist()
```

```python
# Yönsüz Graf
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
nx.draw(G,with_labels=True)
```

```python
#Yönlü Graf
G_directed = nx.DiGraph()
G_directed.add_nodes_from(nodes)
G_directed.add_edges_from(edges)
nx.draw(G_directed,with_labels=True)
```

```python
# Düğüm dereceleri
toplam=0
for node_name in nodes:
  print(f"{node_name} düğümünün derecesi: {G.degree(node_name)}")
  toplam+=G.degree(node_name)
print(f"Ortalama derece: {toplam/len(nodes)}")
```

```python
# Düğüm derecelerinin histogramı (Histogram of node degrees)
degrees = [G.degree(node) for node in G.nodes()]

plt.figure(figsize=(10, 6))
plt.hist(degrees, bins=range(min(degrees), max(degrees) + 2), edgecolor='black', alpha=0.7)
plt.title('Düğüm Derecelerinin Dağılımı (Distribution of Node Degrees)')
plt.xlabel('Düğüm Derecesi (Node Degree)')
plt.ylabel('Düğüm Sayısı (Number of Nodes)')
plt.xticks(range(min(degrees), max(degrees) + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Degree Centrality
degree_centrality = nx.degree_centrality(G)
sorted_degree_centrality = sorted(degree_centrality.items(), key=lambda item: item[1], reverse=True)
for node, centrality in sorted_degree_centrality:
    print(f"Node {node}: Degree Centrality = {centrality}")
```

```python
#Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)
sorted_closeness_centrality = sorted(closeness_centrality.items(), key=lambda item: item[1], reverse=True)
for node, centrality in sorted_closeness_centrality:
    print(f"Node {node}: Closeness Centrality = {centrality}")
```

```python
#Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
sorted_betweenness_centrality = sorted(betweenness_centrality.items(), key=lambda item: item[1], reverse=True)
for node, centrality in sorted_betweenness_centrality:
    print(f"Node {node}: Betweenness Centrality = {centrality}")
```

```python
#EigenVector Centrality
eigenvector_centrality = nx.eigenvector_centrality(G)
sorted_eigenvector_centrality = sorted(eigenvector_centrality.items(), key=lambda item: item[1], reverse=True)
for node, centrality in sorted_eigenvector_centrality:
    print(f"Node {node}: EigenVector Centrality = {centrality}")
```

```python
#Katz Centrality
katz=nx.katz_centrality(G,alpha=0.1) # Reduced alpha to avoid convergence issues
sorted_katz = sorted(katz.items(), key=lambda item: item[1], reverse=True)
for node, katz in sorted_katz:
    print(f"Node {node}: Katz = {katz}")
```

```python
#Pagerank
pagerank=nx.pagerank(G,alpha=0.85)
sorted_pagerank = sorted(pagerank.items(), key=lambda item: item[1], reverse=True)
for node, pagerank in sorted_pagerank:
    print(f"Node {node}: Pagerank = {pagerank}")
```

```python
from networkx.algorithms import community

# Modularity hesaplama
communities = list(community.label_propagation_communities(G))
modularity_score = community.modularity(G, communities)
print(f"G network için Modülarite: {modularity_score}")
```

```python
# G network için density hesaplama
density = nx.density(G)
print(f"G network için Density: {density}")
```

```python
# G network için average clustering hesaplama
average_clustering = nx.average_clustering(G)
print(f"G network için Average Clustering: {average_clustering}")
```

```python
# G network için diameter hesaplama
diameter = nx.diameter(G)
print(f"G network için Diameter: {diameter}")
```

```python
# Düğüm derecelerinin histogramı (Histogram of node degrees)
degrees = [G_directed.degree(node) for node in G_directed.nodes()]

plt.figure(figsize=(10, 6))
plt.hist(degrees, bins=range(min(degrees), max(degrees) + 2), edgecolor='black', alpha=0.7)
plt.title('Düğüm Derecelerinin Dağılımı (Distribution of Node Degrees)')
plt.xlabel('Düğüm Derecesi (Node Degree)')
plt.ylabel('Düğüm Sayısı (Number of Nodes)')
plt.xticks(range(min(degrees), max(degrees) + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# En kritik 5 köprü düğüm(BETWEENNESS)
for node, centrality in sorted_betweenness_centrality[:5]:
    print(f"Node {node}: Betweenness Centrality = {centrality}")
```

```python
#En kritik 5 lider düğüm (EIGENVECTOR)
for node, centrality in sorted_eigenvector_centrality[:5]:
    print(f"Node {node}: EigenVector Centrality = {centrality}")
```

```python
# Güçlü düğümler ve çıkarılması (degree_centrality)
for node, centrality in sorted_degree_centrality[:3]:
    print(f"Node {node}: Degree Centrality = {centrality}")

G.remove_nodes_from(["s01","s03","s04"])
nx.draw(G, with_labels=True)
```
