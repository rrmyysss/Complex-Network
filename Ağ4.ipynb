```python
import networkx as nx
```

```python
# average_clustering
#1. Karate Kulübü Grafiğini Yükleme
G_karate = nx.karate_club_graph()
print(f"Düğüm Sayısı (Üyeler): {G_karate.number_of_nodes()}")
print(f"Kenar Sayısı (İlişkiler): {G_karate.number_of_edges()}")
print("-" * 50)
# 2. Bireysel Kümeleme Katsayılarını Hesaplama (C_i)
individual_coeffs = nx.clustering (G_karate)
# En yüksek ve en düşük C_i değerine sahip düğümleri bulalım
max_node = max(individual_coeffs, key=individual_coeffs.get)
min_node = min(individual_coeffs, key=individual_coeffs.get)
print(f" En Yüksek Ci (Düğüm (max_node)): (individual_coeffs [max_node]:.4f)")
print(f" En Düşük Ci (Düğüm (min_node)): {individual_coeffs [min_node]:.4f}")
print("-" * 50)
#3. Ortalama Kümeleme Katsayısını Hesaplama (C)
avg_cluster_coeff = nx.average_clustering(G_karate)
print(f" Ortalama Kümeleme Katsayısı (C): (avg_cluster_coeff:.4f)")
print("-" * 50)
```

```python
# Diameter
#Karate Kulübü Grafiğini Yükleme
G_karate = nx.karate_club_graph()
print(f"Düğüm Sayısı (Üyeler): (G_karate.number_of_nodes)")
print(f"Kenar Sayısı (İlişkiler): {G_karate.number_of_edges()}")
print("-" * 50)
# Çapı Hesaplama
try:
  diameter = nx.diameter(G_karate)
  print(f"Karate Kulübü Ağının Çapı (Diameter): (diameter)")
except nx.NetworkXError as e:
  # Bu veri seti bağlı olduğu için bu hata beklenmez.
  print(f"Hata: Grafik bağlı değil. Çap hesaplanamaz. {e}")
```

```python
from networkx.algorithms.community import modularity as nx_modularity
#1. Karate Kulübü Grafiğini Yükleme
G_karate = nx.karate_club_graph()
# 2. Tarihsel Olarak Bilinen İki Fraksiyonlu Bölüntü (Partition)
# Düğüm Numaraları (0'dan 33'e kadar)
# Faction 0: Kulüp Müdürü (John A.) Tarafı
community_0 = {0, 1, 2, 3, 7, 11, 12, 13, 17, 19, 21, 30}
# Faction 1: Eğitmen (Mr. Hi) Tarafı (Geri kalan tüm düğümler)
community_1 = set(G_karate.nodes())-community_0
# NetworkX Modularity fonksiyonu, toplulukları bir liste veya küme listesi olarak bekler
karate_partition = [community_0, community_1]
#3. Modülerliği Hesaplama
#nx.modularity (G, communities) fonksiyonu kullanılır
Q_score = nx_modularity(G_karate, karate_partition)
print(f"Düğüm Sayısı: (G_karate.number_of_nodes())")
print(f"Bölüntüdeki Topluluk Sayısı: {len (karate_partition)}")
print("-" * 50)
print(f" Bilinen Bölüntünün Modülerlik Skoru (Q): (Q_score:.4f)")
```
